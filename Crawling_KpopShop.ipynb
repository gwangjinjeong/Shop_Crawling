```python
for category in categories:
  print(category)
  baseurl = "https://www.a-kpop.com"
  url = baseurl+category
  print(url)
  res = urllib.request.urlopen(url)
  soup = BeautifulSoup(res, "html.parser")
  lastpage = int(soup.findAll('a',attrs={'class':'page-number'})[-2].text)
  for i in range(1,lastpage+1):
  # for i in range(1,1+1):
    values = f"/page/{i}"
    url = baseurl + category + values
    res = urllib.request.urlopen(url)
    soup = BeautifulSoup(res, "html.parser")

    for contents in soup.findAll('div', attrs={'class':'product-small box '}):
      title = contents.find('p', attrs={'class',"name product-title woocommerce-loop-product__title"}).text
      link = contents.find(href = True)['href']
      a = contents.findAll('span', attrs={'class',"price"})
      try:
        first = a[0].findAll('span', attrs={'class',"woocommerce-Price-amount amount"})[0].text
        second = a[0].findAll('span', attrs={'class',"woocommerce-Price-amount amount"})[1].text
        origin = first + '-' + second
      except IndexError:
        try:
          origin = a[0].findAll('span', attrs={'class',"woocommerce-Price-amount amount"})[0].text
        except IndexError:
          origin = None
      try:
        origin = a[0].find('del').text
        sale = second
      except (IndexError,AttributeError):
        sale = origin
      tmpimg = contents.find('div',attrs={'class':'image-none'})
      img = tmpimg.find('img')['src']
      items = items.append(pd.Series([link,title,sale,origin,img], index=items.columns), ignore_index=True)
```
